```python
from google.colab import files
import os
import subprocess
from IPython.display import HTML, display
import shutil

def reverse_video_con_audio(input_path, output_path):
   
    try:
        cmd = [
            'ffmpeg',
            '-i', input_path,
            '-vf', 'reverse',
            '-af', 'areverse',
            '-y',
            output_path
        ]

        print(f"Procesando: {os.path.basename(input_path)}")
        result = subprocess.run(cmd, capture_output=True, text=True)

        if result.returncode == 0:
            print(f"Video generado: {os.path.basename(output_path)}")
            return True
        else:
            print(f"Error: {result.stderr[:200]}...")
            return False

    except Exception as e:
        print(f"Error procesando: {str(e)}")
        return False


def mostrar_video(video_path):
    
    from base64 import b64encode
    video_file = open(video_path, "rb").read()
    data_url = "data:video/mp4;base64," + b64encode(video_file).decode()
    return HTML(f"""
    <video width="400" controls>
        <source src="{data_url}" type="video/mp4">
    </video>
    <p>{os.path.basename(video_path)}</p>
    """)


def main():
    shutil.rmtree('/content/videos_originales', ignore_errors=True)
    shutil.rmtree('/content/videos_revertidos', ignore_errors=True)

    os.makedirs('/content/videos_originales', exist_ok=True)
    os.makedirs('/content/videos_revertidos', exist_ok=True)


    print("SUBE TUS VIDEOS (puedes seleccionar mÃºltiples)")
    uploaded = files.upload()

    if not uploaded:
        print("No se subieron archivos")
        return

    for filename in uploaded.keys():
        shutil.move(filename, f'/content/videos_originales/{filename}')

    print(f"Videos subidos: {list(uploaded.keys())}")
    print()

    print("REVIRTIENDO VIDEOS...")
    for filename in uploaded.keys():
        input_path = f'/content/videos_originales/{filename}'
        name, ext = os.path.splitext(filename)
        output_path = f'/content/videos_revertidos/{name}.mp4'

        reverse_video_con_audio(input_path, output_path)

    print()

    print("DESCARGAR VIDEOS REVERTIDOS")

    shutil.make_archive('videos_revertidos', 'zip', '/content/videos_revertidos')

    print("Descargando archivo ZIP con todos los videos revertidos...")
    files.download('videos_revertidos.zip')

    print("Â¡Proceso completado!")


if __name__ == "__main__":
    main()
```
